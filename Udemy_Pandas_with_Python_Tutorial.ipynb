### filtrando dados

```python
booleans = []
for lenght in data.duration:
    if lenght >= 175:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
data[booleans]
```

```python
data[data.duration >= 175].genre
```

#### filtrar varias condições

```python
data[(data.content_rating == 'R') & (data.genre == 'Crime')]
```

```python
data[(data.genre == 'Crime') | (data.genre == 'Action') | (data.genre == 'Drama')]
```

```python
data[data.genre.isin(['Crime','Action','Drama'])]
```

### group by functions

```python
data = pd.read_csv('http://bit.ly/imdbratings', dtype = {'duration':int,'star_rating':float})
```

```python
data.head(5)
```

```python
data.groupby('genre').duration.mean().sort_values(ascending = False)
```

```python
data.groupby('genre').duration.agg(['min','median','mean','max','count'])
```
